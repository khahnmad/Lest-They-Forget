## How often are each of the shootings mentioned?
Looking at the mention of the shooting name (a couple versions of the name) and the shooter's name

```python
# IMPORTS
import All_Functions as af
import pandas as pd
import ast
from nltk import FreqDist
import glob
```

### Create the frequency dataframe
It is not necessary to run this section because the later part of this notebook just imports the csv file "v2-sample_shootins.csv" which is generated from this section of code  

```python
# FUNCTIONS
def shooting_term_frequency(df, location, shooting_name, shooter_name):
    df_applied = af.apply_ngrams(df) # create tokens, bigrams, and trigrams 
    
    # Check for the frequency of the following words
    af.freq_analysis(df_applied,shooting_name)
    af.freq_analysis(df_applied, 'shooting')
    af.freq_analysis(df_applied,shooter_name)
#     print(df_applied)
    summary = [location, df_applied[shooting_name[:3]].sum(),df_applied['shooting'[:3]].sum(), df_applied[shooter_name[:3]].sum()]
    return summary
```

```python
all_files = [x for x in glob.glob('newspaper-text' + "/*.csv") if "v2" in x]
```

```python
summary = [['City Or County', 'Shooting Name Sum', 'Shooting Sum','Shooter Name Sum']]
# San Bernardino 
print('Getting San Bernardino...')
sanbernardino_df = af.combine_months_to_df([x for x in all_files if 'SanBern' in x])
summary.append(shooting_term_frequency(sanbernardino_df,'San Bernardino', 'san bernardino shooting', 'farook')) 
""" There are two shooters: farook and malik. Farook (the husband) is mentioned more frequently than Malik so he's the name I'm including """

# Houston 
print('Getting Houston...')
houston_df = af.combine_months_to_df([x for x in all_files if 'Houston' in x])
summary.append(shooting_term_frequency(houston_df,'Houston', 'falling oaks', 'torres')) 

# DC
print('Getting DC...')
dc_df = af.combine_months_to_df([x for x in all_files if 'DC' in x])
summary.append(shooting_term_frequency(dc_df,'Washington Navy Yard', 'navy yard shooting', 'alexis'))

# Boulder 
print('Getting Boulder...')
boulder_df = af.combine_months_to_df([x for x in all_files if 'Boulder' in x])
summary.append(shooting_term_frequency(boulder_df,'Boulder', 'boulder shooting', 'ahmad'))
"""The shooter's full name is Ahmad Al Aliwi Al-Issa. ahmad gets 357 hits, aliwi gets 337 hits, other combinations get very few hits. So I'm using 'ahmad' as the search variable """

# Odessa
print('Getting Odessa...')
odessa_df = af.combine_months_to_df([x for x in all_files if 'Odessa' in x])
summary.append(shooting_term_frequency(odessa_df,'Odessa', 'odessa shooting', 'ator'))

# Pittsburgh
print('Getting Pittsburgh...')
pittsburgh_df = af.combine_months_to_df([x for x in all_files if 'Pitts' in x])
summary.append(shooting_term_frequency(pittsburgh_df,'Pittsburgh', 'pittsburgh shooting', 'bowers'))

# Las Vegas
print('Getting Las Vegas...')
vegas_df = af.combine_months_to_df([x for x in all_files if 'Vegas' in x])
vegas_df_ = vegas_df.drop(1090) # get rid of an outlier 
summary.append(shooting_term_frequency(vegas_df_,'Las Vegas', 'las vegas shooting', 'paddock'))

# Plano 
print('Getting Plano...')
plano_df = af.combine_months_to_df([x for x in all_files if 'Plano' in x])
plano_df_ = plano_df.drop(1652) # get rid of an outlier 
summary.append(shooting_term_frequency(plano_df_,'Plano', 'plano shooting', 'hight'))
"""The shooter, Spencer Hight, also killed his wife with the same last name, which may complicate the results"""

# Bogue Chitto
print('Getting Bogue Chitto...')
bogue_df = af.combine_months_to_df([x for x in all_files if 'Bogue' in x])
summary.append(shooting_term_frequency(bogue_df,'Bogue Chitto', 'bogue chitto shooting', 'godbolt'))

# Virginia Beach 
print('Getting Virginia Beach...')
bogue_df = af.combine_months_to_df([x for x in all_files if 'Virginia' in x])
summary.append(shooting_term_frequency(bogue_df,'Virginia Beach', 'virginia beach shooting', 'craddock'))
```

```python
# get summary of info: positive/ neg sentiment more common, most common topics, las mentions, pad mentions 
def get_newspaper_summary(subset_df):

    # Get the distribution of sentiments
    neg_count, pos_count, neu_count = 0,0,0
    for index, row in subset_df.iterrows():
        if int(subset_df.loc[index, 'Sentiment']['compound']) >0:
            pos_count += 1
        else:
            neg_count += 1

#     print(subset_df.columns)
#     # Get the sum of shooting mentions and shooter name mentions
#     print(f'Shooting Name mention {subset_df.iloc[:,[11]].sum()}')
#     print(f'Shooter Name mention {subset_df.iloc[:,[12]].sum()}')

    # Get the most common themes
    themes = []
    for index, row in subset_df.iterrows():
        if type(subset_df.loc[index,'themes']) == str:  
            instance = subset_df.loc[index,'themes'].split(',')
            for item in instance:
                themes.append(item)
    dict_  = {}
    for x in themes:
        dict_[x] = themes.count(x)

    new = FreqDist(dict_)
    most_common = new.most_common(5)
#     print(f'Most common themes: {most_common}')
    return [len(subset_df), [neg_count, pos_count], subset_df.iloc[:,11].sum(),subset_df.iloc[:,12].sum(),most_common]
```

```python
def newspaper_summary_pipeline(df, shooting_name, shooter_name):
    df_applied = af.apply_ngrams(df)
    af.freq_analysis(df_applied,shooting_name)
    af.freq_analysis(df_applied, shooter_name)
    
    sentiment =df["sentiment"].astype('str')
    sentiment = sentiment.apply(lambda x: ast.literal_eval(x))
    df['Sentiment'] = sentiment
    df=df.drop(columns=['sentiment'])

    right_df = df[df.loc[:,'Leaning'] == 'right']
    center_right_df = df[df.loc[:,'Leaning'] == 'center-right']
    center_left_df = df[df.loc[:,'Leaning'] == 'Center Left']
    center_df = df[df.loc[:,'Leaning'] == 'center']
    left_df = df[df.loc[:,'Leaning'] == 'left']
    
    data = []
    
    data.append(get_newspaper_summary(right_df))
    data.append(get_newspaper_summary(center_right_df))
    data.append(get_newspaper_summary(center_df))
    data.append(get_newspaper_summary(center_left_df))
    data.append(get_newspaper_summary(left_df))

    summary_df =pd.DataFrame(data,columns=['Number of articles','Sentiment Count','Shooting Name','Shooter Name','Most common themes'])
    summary_df['Leaning'] = ['Right','Center Right','Center','Center Left', 'Left']
    return summary_df
```
