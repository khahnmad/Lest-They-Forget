```python
# IMPORTS
import All_Functions as af
import nltk
from collections import Counter
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import glob
```

```python
def add_name_to_df(df, shooting_name):
    for index, row in df.iterrows():
        df.loc[index,'Shooting Name'] = shooting_name

def get_text_freq(subset):
    all_right_text = [x.split(' ') for x in subset['text']]
    text = []
    for item in all_right_text:
        for token in item:
            text.append(token)
#     return Counter(text).most_common()
    return text

def get_date_numbers(df):
    df['publish_date']=pd.to_datetime(df['publish_date'])
    min_value = df.sort_values(by=['publish_date'], ascending=True).head(1)
    min_value = min_value.reset_index(drop=True)
    for index, row in df.iterrows():
        date_number = df.loc[index,'publish_date'] - min_value.loc[0,'publish_date']
        string = str(date_number).split()
        if string[0] != "NaT":
            new = int(string[0])
            df.loc[index, 'date_number'] = new
```

## Does one political leaning talk about mental illness more than the others ?

```python
mental_single = [
'psychiatry'
,'troubled'
,'isolated'
,'anxiety'
,'depression'
,'psychiatrist'
,'psychologist'
,'psychotic'
,'paranoid'
,'delusional'
, 'psychotropic','psychiatrists','pyschiatric']
mental_bigrams = ['mental hospital','hearing voices'
,'mental illness'
,'mental health']
mental_trigrams = ['mental health facilities','post-traumatic stress disorder']
```

```python
def count_dict_freq(data):
    count = 0
    for x in data:
        if x in mental_single:
            count += 1
    bigrams = af.extract_ngrams(data, 2)
    for x in bigrams:
        if x in mental_bigrams:
            count += 1
    trigrams = af.extract_ngrams(data, 3)
    for x in trigrams:
        if x in mental_trigrams:
            count += 1
    return count 
```

```python
print('MENTAL HEALTH DISCUSSION')
print(f'Right: {count_dict_freq(right_freq)}, {count_dict_freq(right_freq)/len(right_freq)}')
print(f'Center right: {count_dict_freq(centerright_freq)}, {count_dict_freq(centerright_freq)/len(centerright_freq)}')
print(f'Center: {count_dict_freq(center_freq)}, {count_dict_freq(center_freq)/len(center_freq)}')
print(f'Center left: {count_dict_freq(centerleft_freq)}, {count_dict_freq(centerleft_freq)/len(centerleft_freq)}')
print(f'Left: {count_dict_freq(left_freq)}, {count_dict_freq(left_freq)/len(left_freq)}')
```
