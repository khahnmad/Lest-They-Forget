```python
# IMPORTS
import All_Functions as af
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def get_date_numbers(df):
    df['publish_date']=pd.to_datetime(df['publish_date'])
    min_value = df.sort_values(by=['publish_date'], ascending=True).head(1)
    min_value = min_value.reset_index(drop=True)
    for index, row in df.iterrows():
        date_number = df.loc[index,'publish_date'] - min_value.loc[0,'publish_date']
        string = str(date_number).split()
        if string[0] != "NaT":
            new = int(string[0])
            df.loc[index, 'date_number'] = new
            
def shooting_term_intensity(df, location, shooting_name, shooter_name):
    df_applied = af.apply_ngrams(df) # create tokens, bigrams, and trigrams 
    
    # Check for the frequency of the following words
    af.freq_analysis(df_applied,shooting_name)
    af.freq_analysis(df_applied, 'shooting')
    af.freq_analysis(df_applied,shooter_name)
    
    # Create standardized numbers for the three-month timeline
    get_date_numbers(df_applied)
    
#     print('Max:',df_applied['date_number'].max())
    intensity_summary = [['City Or County','Day range','Shooting name','Shooting','Shooter Name']]
    for x in range(1,10):
        upper_limit = x * 11 
        lower_limit = upper_limit - 11
     
        eleven_days = df_applied[df_applied['date_number']< upper_limit] 
        lower=  eleven_days[eleven_days['date_number']>=lower_limit] # should have a lower limit
        
        shootingname = (lower[shooting_name[:3]].sum() - 0) / (299-0)
        shooting = (lower['shooting'[:3]].sum()-0)/(3661-0)
        shootername = (lower[shooter_name[:3]].sum()-0)/(1985-0)
        
#         shootingname = lower[shooting_name[:3]].sum()
#         shooting = lower['shooting'[:3]].sum()
#         shootername = lower[shooter_name[:3]].sum()
        
        
        intensity_summary.append([location,str(lower_limit)+'-'+str(upper_limit),shootingname, shooting, shootername])
    
    df = pd.DataFrame(data=intensity_summary[1:], columns=intensity_summary[0])
    print('Shooting name',df['Shooting name'].max(), df['Shooting name'].min())
    print('Shooting ',df['Shooting'].max(), df['Shooting'].min())
    print('Shooter Name',df['Shooter Name'].max(), df['Shooter Name'].min())
#     print(df)
    return df
```

```python
summary = [['City Or County','Day range','Shooting name','Shooting','Shooter Name']]
# San Bernardino 
print('Getting San Bernardino...')
sanbernardino_df = af.combine_months_to_df('newspaper-text\SanBernadino_Text-first-month.csv','newspaper-text\SanBernadino_Text-second-month.csv','newspaper-text\SanBernadino_Text-third-month.csv')
sanbern_intensity = shooting_term_intensity(sanbernardino_df,'San Bernardino', 'san bernardino shooting', 'malik')
# print(sanbern_intensity)
# greatest = 0
# for i in range(1,len(sanbern_intensity)):
#     print(row)
#     if int(sanbern_intensity[i][2]) > greatest:
#         greatest = int(sanbern_intensity[i][2])
    

# # Houston 
print('Getting Houston...')
houston_df = af.combine_months_to_df('newspaper-text\Houston_Text-first-month.csv','newspaper-text\Houston_Text-second-month.csv','newspaper-text\Houston_Text-third-month.csv')
houston_intensity = shooting_term_intensity(houston_df,'Houston', 'falling oaks', 'torres')

# # DC
print('Getting DC...')
dc_df = af.combine_months_to_df('newspaper-text\DC_Text-first-month.csv','newspaper-text\DC_Text-second-month.csv','newspaper-text\DC_Text-third-month.csv')
dc_intensity = shooting_term_intensity(dc_df,'Washington Navy Yard', 'navy yard shooting', 'alexis')

# # Boulder 
print('Getting Boulder...')
boulder_df = af.combine_months_to_df('newspaper-text\Boulder_Text-first-month.csv','newspaper-text\Boulder_Text-second-month.csv','newspaper-text\Boulder_Text-third-month.csv')
boulder_intesity = shooting_term_intensity(boulder_df,'Boulder', 'boulder shooting', 'ahmad')

# # Odessa
print('Getting Odessa...')
odessa_df = af.combine_months_to_df('newspaper-text\Odessa_Text-first-month.csv','newspaper-text\Odessa_Text-second-month.csv','newspaper-text\Odessa_Text-third-month.csv')
odessa_intesity = shooting_term_intensity(odessa_df,'Odessa', 'odessa shooting', 'ator')

# Pittsburgh
print('Getting Pittsburgh...')
pittsburgh_df = af.combine_months_to_df('newspaper-text\Pittsburgh_Text-first-month.csv','newspaper-text\Pittsburgh_Text-second-month.csv','newspaper-text\Pittsburgh_Text-third-month.csv')
pittsburgh_intesity= shooting_term_intensity(pittsburgh_df,'Pittsburgh', 'pittsburgh shooting', 'bowers')

# Las Vegas
print('Getting Las Vegas...')
vegas_df = af.combine_months_to_df('newspaper-text/v2-Vegas_Text-first-month.csv','newspaper-text/v2-Vegas_Text-second-month.csv','newspaper-text/v2-Vegas_Text-third-month.csv')
vegas_df_ = vegas_df.drop(1090) # get rid of an outlier 
vegas_intesity = shooting_term_intensity(vegas_df_,'Las Vegas', 'las vegas shooting', 'paddock')

# Plano 
print('Getting Plano...')
plano_df = af.combine_months_to_df('newspaper-text\Plano_Text-first-month.csv','newspaper-text\Plano_Text-second-month.csv','newspaper-text\Plano_Text-third-month.csv')
plano_df_ = plano_df.drop(1652) # get rid of an outlier 
plano_intensity =shooting_term_intensity(plano_df_,'Plano', 'plano shooting', 'hight')

# Bogue Chitto
print('Getting Bogue Chitto...')
bogue_df = af.combine_months_to_df('newspaper-text\Bogue_Text-first-month.csv','newspaper-text\Bogue_Text-second-month.csv','newspaper-text\Bogue_Text-third-month.csv')
bouge_intensity = shooting_term_intensity(bogue_df,'Bogue Chitto', 'bogue chitto shooting', 'godbolt')
```
